```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
data={"color":["Red","Green","Yellow","orange","pink"]}
df=pd.DataFrame(data)
df
```

```python
label_encoder=LabelEncoder()
df['Color']=label_encoder.fit_transform(df['color'])
df.head()
```

```python
one_hot_encoder=OneHotEncoder()
one_hot_encoded_data=one_hot_encoder.fit_transform(df[['Color']])
one_hot_encoded=pd.DataFrame(one_hot_encoded_data.toarray(),columns=one_hot_encoder.get_feature_names_out(['Color']))
df=pd.concat([df,one_hot_encoded],axis=1)
df
```
